```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
point_list = [0]
matrix = np.ones((3,3))
choices = ["r","p","s"]
WHAT_WINS = {"r" : "p", "p" : "s", "s" : "r"}

prev_choice = np.random.choice(choices)

while True:
  user_choice = input(f"r - rock, p - paper, s - scissors, q - quit")

  if user_choice == "q":
    break
  if user_choice not in choices:
    print("try again")
    continue
  
  prev_index = choices.index(prev_choice)
  index = choices.index(user_choice)
  row = matrix[prev_index]/matrix[prev_index].sum()
  print(matrix)
  cp_choice = WHAT_WINS[np.random.choice(choices, p=row)]


  if WHAT_WINS[cp_choice] == user_choice:
    point_list.append(point_list[-1]+1)
    matrix[prev_index,index] -= 0.4*matrix[prev_index,index]
    print("Computer Wins")
  elif cp_choice == user_choice:
    point_list.append(point_list[-1])
    print("Tie")
  else:
    point_list.append(point_list[-1]-1)
    matrix[prev_index,index] += 0.2*matrix[prev_index,index]
    print("You Win")
  matrix = normalize(matrix,axis=1,norm='l1')

  prev_choice = user_choice
  
print(matrix)

plt.figure()
plt.plot(point_list, "-x")
plt.show()


  
```
