```python
from google.colab import drive
import cv2
drive.mount('/content/drive')
image1 = cv2.imread("/content/drive/MyDrive/WMA/zd1.jpg")
image2 = cv2.imread("/content/drive/MyDrive/WMA/zd2.jpg")
image3 = cv2.imread("/content/drive/MyDrive/WMA/zd3.jpg")
image4 = cv2.imread("/content/drive/MyDrive/WMA/zd4.jpg")
image5 = cv2.imread("/content/drive/MyDrive/WMA/zd5.jpg")
video = cv2.VideoCapture("/content/drive/MyDrive/WMA/vi.mp4")
```

sift

```python
import cv2
import matplotlib.pyplot as plt
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
gray4 = cv2.cvtColor(image4, cv2.COLOR_BGR2GRAY)
gray5 = cv2.cvtColor(image5, cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(gray1,None)
kp2, des2 = sift.detectAndCompute(gray2,None)
kp3, des3 = sift.detectAndCompute(gray3,None)
kp4, des4 = sift.detectAndCompute(gray4,None)

while(video.isOpened()):
  ret, frame = video.read()
  if ret == True:
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    kp5, des5 = sift.detectAndCompute(gray_frame,None)
    bf1 = cv2.BFMatcher()
    bf2 = cv2.BFMatcher()
    bf3 = cv2.BFMatcher()
    bf4 = cv2.BFMatcher()
    bf5 = cv2.BFMatcher()

    m1 = bf1.knnMatch(des1,des5, k=2)
    m2 = bf1.knnMatch(des2,des5, k=2)
    m3 = bf1.knnMatch(des3,des5, k=2)
    m4 = bf1.knnMatch(des4,des5, k=2)
    list1 = {
        m1 : image1,
        m2 : image2,
        m3 : image3,
        m4 : image4
    }
    sorted = dict(sorted(list1.items()))
    key = next(iter(sorted))
    image = list1(key)
```
