```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
state = {
    "Papier" : 0,
    "Kamień" : 1,
    "Nożyce" : 2
}
state
```

```python
matrix = np.array([[0.3,0.3,0.4],[0.4,0.3,0.3],[0.3,0.4,0.3]])
matrix
```

```python
payoffs = {'Papier Kamień': 1, 'Kamień Nożyce': 1, 'Nożyce Papier': 1,
           'Kamień Papier': -1, 'Nożyce Kamień': -1, 'Papier Nożyce': -1,
           'Papier Papier': 0, 'Kamień Kamień': 0, 'Nożyce Nożyce': 0}
payoffs
```

```python
to_plot = np.array([0])
```

```python
def update_matrix(prev_state, cur_state, learning_rate):
    global matrix
    matrix[state[prev_state]][state[cur_state]] += learning_rate
    row_sums = matrix.sum(axis=1)
    matrix = matrix / row_sums[:, np.newaxis]
    return matrix
```

```python
def computer_move():
  return random.choice(list(state.values()))
```

```python
def play_round():
    comp1 = computer_move()
    comp2 = computer_move()
    payoff = payoffs[comp1 + " " + comp2]
    np.append(to_plot,np.array(payoff + to_plot[-1]))
    return comp2
```

```python
def play_game(rounds,rate):
    current_state = random.choice(list(state.values()))
    for i in range(rounds):
      previous_state = current_state
      current_state = play_round()
      transition_matrix = update_matrix(previous_state,current_state,rate)
    return transition_matrix
```

```python
matrix = play_game(30,0.1)
```
